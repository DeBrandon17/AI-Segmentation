```python
#importing the necesscary libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler,normalize
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn import metrics
data=pd.read_csv("/content/cell2celltrain.csv")
data.head()
data.describe()
data.info()
```

```python
#data validation
data.head()
data.dropna()
print(data)
```

```python
#this code turns the Yes and No into boolean data which can be proccessed
data=data[['Churn','MonthlyRevenue','TotalRecurringCharge','DirectorAssistedCalls','OverageMinutes','RoamingCalls','DroppedCalls','ThreewayCalls','ReceivedCalls','OutboundCalls','InboundCalls','MonthsInService','BuysViaMailOrder']]
from sklearn import preprocessing
k=preprocessing.LabelEncoder()
k.fit(data['Churn'])
data['Churn']=k.transform(data['Churn'])
data.dropna(inplace=True)
data=data[['Churn','MonthlyRevenue','TotalRecurringCharge','DirectorAssistedCalls','OverageMinutes','RoamingCalls','DroppedCalls','ThreewayCalls','ReceivedCalls','OutboundCalls','InboundCalls','MonthsInService','BuysViaMailOrder']]

k=preprocessing.LabelEncoder()
k.fit(data['BuysViaMailOrder'])
data['BuysViaMailOrder']=k.transform(data['BuysViaMailOrder'])
data.dropna(inplace=True)
data=data[['Churn','MonthlyRevenue','TotalRecurringCharge','DirectorAssistedCalls','OverageMinutes','RoamingCalls','DroppedCalls','ThreewayCalls','ReceivedCalls','OutboundCalls','InboundCalls','MonthsInService','BuysViaMailOrder']]
```

K means is a type of unsupervised learning which involves putting unlabelled data into clusters.

1. Is to import the libraries needed, this is a common step for all types of algorithims.
2. Then we need to read the data set using pandas llibrary , and assign the data into a variable
3. We have to explore the data using data info seeing what is the data we are dealing with the amount of columns, and rows, also the max data the median data and the minumum data
4. After that we need to make any boolean data into 1 and 0's using label encoding
5. Then we need to plot the data into a graph to see any correlation, over here we are using a heatmap to determine the correlation
6. Then we need to standardize the data so its in the range of 0 and 1 so they all are in the same range
7.  Then we have to deternine how many clusters should be in the algorithm, we do this using WCSS, and then we need to graph it, this is known as elbow method, where the bend appears is the number of clusters needed
8. Then we fit our model to our specific dataset
9. Finally we identify what are the clusters sepretaed based of by seeing the average values for each cluster and parapeters


```python
#plots  hearmap to show the correlation betwen data
plt.figure(figsize=(12,9))
sns.heatmap(data.corr(),annot=True,cmap='RdBu')
plt.title('Correlation Heatmap',fontsize=14)
plt.yticks(rotation =0)
plt.show()
```

```python
#standardises the data
scaler = StandardScaler()
data_std = scaler.fit_transform(data)
data_std = pd.DataFrame(data = data_std,columns = data.columns)
```

```python
#fits pca with our dataset
pca = PCA(n_components = 3)
pca.fit(data_std)
pca.components_
pca.transform(data_std)
scores_pca = pca.transform(data_std)
```

```python
#used to find the amount of clusters
wcss = []
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
#the elbow value is picked as number of clusters  
plt.figure(figsize = (10,8))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '-.',color='red')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

```python
#segments the data using the optimal number of clusters using kmeans
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
kmeans.fit(data_std)
data_segm_kmeans= data_std.copy()
data_segm_kmeans = pd.DataFrame(data = data_std,columns = data.columns)
data_segm_kmeans["what"] = kmeans.labels_
```

```python
#analyses the data
data_segm_analysis = data_segm_kmeans.groupby(["what"]).mean()
data_segm_analysis.head()
```

1st segment less monthly revenue with less calls
2nd semgent high revenue with high calls

```python
#names each segment 
data_segm_analysis.rename({0:'Less revenue/calls',
                          1:'Average',
                          2:'High revenue/calls',
                          3:'BuysViaMailorder,high'})
data_segm_kmeans['Labels'] = data_segm_kmeans["what"].map({0:'Less revenue/calls',
                                                          1:'Average',
                                                          2:'High revenue/calls',
                                                          3:'BuysViaMailorder,high'})
```

```python
#plots the segmented data
x_axis = data_segm_kmeans['MonthlyRevenue']
y_axis = data_segm_kmeans['TotalRecurringCharge']
plt.figure(figsize = (10, 8))
sns.scatterplot(x_axis, y_axis, hue = data_segm_kmeans['Labels'], palette = ['g', 'r','b','y'])
plt.title('Segmentation K-means')
plt.show()
```

```python
#standardises and fits the data to be used by PCA
pca = PCA()
data_std = scaler.fit_transform(data)
data_std = pd.DataFrame(data = data_std,columns = data.columns)
pca.fit(data_std)
pca.explained_variance_ratio_
```

```python
#Plots the explained variance by components
plt.figure(figsize = (17,9))
plt.plot(range(1,14), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
#fits the data using the best number of components
pca = PCA(n_components = 7)
pca.fit(data_std)
pca.components_
data.columns
```

```python
#names the components
data_pca_comp = pd.DataFrame(data = pca.components_,
                           columns = data.columns,
                           index = ['Component 1', 'Component 2', 'Component 3','Comp 4','Comp 5','Comp 6','Comp 7'])
data_pca_comp
```

```python
#heatmap to show correlation between components and original features
plt.figure(figsize=(12,9))
sns.heatmap(data_pca_comp,
            vmin = -1, 
            vmax = 1,
            cmap = 'RdBu',
            annot = True)
plt.yticks([0, 1, 2, 3, 4, 5, 6,], 
           ['Component 1', 'Component 2', 'Component 3','Comp 4','Comp 5','Comp 6','Comp 7'],
           rotation = 45,
           fontsize = 12)
plt.title('Components vs Original Features',fontsize = 14)
plt.show()
```

```python
#normalises the data
norm=normalize(data_std)
data_norm=pd.DataFrame(norm)
```

```python
#used to get the silhouette score for different number of clusters and find which is the best
for i in range(4,11):
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(data_norm)
    print("Silhouette score for {} clusters k-means : {} ".format(i, metrics.silhouette_score(data_norm,kmeans_labels, metric='euclidean').round(3)))
```

```python
#used to get the Davies Bouldin score for different number of clusters and find which is the best
for i in [4,5,6]:
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(data_norm)
    print('Davies Bouldin Score:'+str(metrics.davies_bouldin_score(data_norm,kmeans_labels).round(3)))
```
